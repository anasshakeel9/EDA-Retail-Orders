```python
import zipfile
zip_ref = zipfile.ZipFile('orders.csv.zip')
zip_ref.extractall() #extract file to dir 
zip_ref.close() #close file
```

```python
import pandas as pd 
df = pd.read_csv('orders.csv',na_values=(['Not Available','unknown']))
```

```python
df.head()
df['Ship Mode'].unique()
```

```python
#df.rename(columns={'Order Id' : 'order_id', 'City' : 'city'})
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace( " ","_")
df.columns
```

```python
#derive new columns discount, sale price , profit
df['discount']=df['list_price']*df['discount_percent']*.01
df['sale_price']=df['list_price']-df['discount']
df['profit'] = df['sale_price']-df['cost_price']
df
```

```python
#change datatype
df['order_date']=pd.to_datetime(df['order_date'],format='%Y-%m-%d')
df.dtypes
```

```python
 #drop cost_price, list_price and discount_percent columns  
df.drop(columns=['list_price','cost_price','discount_percent'],inplace=True)
df
```

```python
#import data into mysql database
import sqlalchemy as sal

engine = sal.create_engine('mysql+mysqlconnector://root:root@localhost/master')
conn = engine.connect()
```

```python
df.to_sql('df_orders', con=conn, index = False, if_exists = 'append')
```
